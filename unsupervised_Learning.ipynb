```
from sklearn.cluster import KMeans
import numpy as np
X=np.array([[1,2],[1,3],[1,4],[2,2],[2,3],[2,4],[3,3],[3,4],[3,5],[4,4]])
kmean=KMeans(n_clusters=4,random_state=0).fit(X)
print(kmean.labels_)
print(kmean.predict([[2,1],[3,6]]))
print(kmean.cluster_centers_)
print(kmean.inertia_)  ## Standard Squared Error (SSE)
```

```
import matplotlib.pyplot as plt

plt.scatter(X[:,0],X[:,1],c=kmean.labels_.astype(float))
plt.show()
```

```
K=range(1,10)
inertias=[]
mapping=[]
for k in K:
  kmeans=KMeans(n_clusters=k,random_state=0).fit(X)
  inertias.append(kmeans.inertia_)

for i in inertias:
  print(i)
```

```
plt.plot(K,inertias)
```

```
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_moons
```

```
X,Y=make_moons(n_samples=1000,noise=0.05)
dbscan=DBSCAN(eps=0.20, min_samples=5)
dbscan.fit(X)
```

```
plt.scatter(X[:,0],X[:,1],c=dbscan.labels_.astype(float))
plt.show()
```
